## **2.1 Descomposición de la Serie Temporal sin Dependencia del Tiempo**

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Descomposición
descomposicion = seasonal_decompose(df_22_23['Daily AQI Value'], model='additive', period=30)

# Graficar la descomposición
fig, ax = plt.subplots(4, 1, figsize=(18, 12), sharex=True)

# Observado
ax[0].plot(range(len(df_22_23)), descomposicion.observed, color='b', label='Observado')
ax[0].set_title('Descomposición de la Serie Temporal', fontsize=17)
ax[0].legend()
ax[0].grid(True, linestyle='--', alpha=0.7)

# Tendencia
ax[1].plot(range(len(df_22_23)), descomposicion.trend, color='r', label='Tendencia')
ax[1].legend()
ax[1].grid(True, linestyle='--', alpha=0.7)

# Estacionalidad
ax[2].plot(range(len(df_22_23)), descomposicion.seasonal, color='g', label='Estacionalidad')
ax[2].legend()
ax[2].grid(True, linestyle='--', alpha=0.7)

# Residuos
ax[3].plot(range(len(df_22_23)), descomposicion.resid, color='k', label='Residuos')
ax[3].legend()
ax[3].grid(True, linestyle='--', alpha=0.7)

# Ajustar las etiquetas del eje X
for a in ax:
    a.set_xticks(range(0, len(df_22_23), max(1, len(df_22_23)//10)))
    a.set_xticklabels(df_22_23.index[::max(1, len(df_22_23)//10)].strftime('%Y-%m-%d'), rotation=60, fontsize=10)

plt.xlabel('Número de Observación', fontsize=15)
plt.tight_layout()
plt.show()


print(f'\nEl residuo cuando la periosidad tenemos una periocidad de 7 es {descomposicion.resid.mean()}')
```

### 2.1.1 Modelo "additive"

```python
# Lista de periodicidades a evaluar
periodicidades = [7, 15, 21, 30, 37, 44, 50]

# Diccionario para almacenar los resultados
residuos_medios_additive = {}

# Evaluar cada periodicidad
for periodo in periodicidades:
    # Descomposición de la serie temporal
    descomposicion = seasonal_decompose(df_22_23['Daily AQI Value'], model='additive', period=periodo)

    # Obtener los residuos
    residuos = descomposicion.resid.mean()
    # Guardar el resultado
    residuos_medios_additive[periodo] = residuos

# Encontrar la periodicidad con la menor media de residuos
mejor_periodicidad = min(residuos_medios_additive, key=residuos_medios_additive.get)


print("Media absoluta de los residuos para cada periodicidad:")
for periodo, media in residuos_medios_additive.items():
    print(f"Periodo: {periodo} - Media de Residuos: {media}")

print(f"\nLa periodicidad ideal basada en la menor media de residuos es: {mejor_periodicidad}")
```

### 2.1.2 Modelo "multiplicative"

```python
# Lista de periodicidades a evaluar
periodicidades = [7, 15, 21, 30, 37, 44, 50]

# Diccionario para almacenar los resultados
residuos_medios_multiplicative = {}

# Evaluar cada periodicidad
for periodo in periodicidades:
    # Descomposición de la serie temporal
    descomposicion = seasonal_decompose(df_22_23['Daily AQI Value'], model='multiplicative', period=periodo)

    # Obtener los residuos
    residuos = descomposicion.resid.mean()
    # Guardar el resultado
    residuos_medios_multiplicative[periodo] = residuos

# Encontrar la periodicidad con la menor media de residuos
mejor_periodicidad = min(residuos_medios_multiplicative, key=residuos_medios_multiplicative.get)


print("Media absoluta de los residuos para cada periodicidad:")
for periodo, media in residuos_medios_multiplicative.items():
    print(f"Periodo: {periodo} - Media de Residuos: {media}")

print(f"\nLa periodicidad ideal basada en la menor media de residuos es: {mejor_periodicidad}")
```

Dado que ambos modelos tienen la menor media de residuos en el periodo de 7, nos basaremos en elegir el modelo aditivo porque a amplitud de la estacionalidad es constante.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Graficar ACF
plt.figure(figsize=(15, 6))
plot_acf(df_22_23['Daily AQI Value'], lags=40, ax=plt.gca())
plt.title('Función de Autocorrelación (ACF)', fontsize=17)
plt.xlabel('Lags', fontsize=15)
plt.ylabel('Autocorrelación', fontsize=15)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Graficar PACF
plt.figure(figsize=(15, 6))
plot_pacf(df_22_23['Daily AQI Value'], lags=40, ax=plt.gca())
plt.title('Función de Autocorrelación Parcial (PACF)', fontsize=17)
plt.xlabel('Lags', fontsize=15)
plt.ylabel('Autocorrelación Parcial', fontsize=15)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Aplicar la prueba de Dickey-Fuller aumentada (ADF)
result = adfuller(df_22_23['Daily AQI Value'])

# Extraer los resultados
print("Estadístico ADF:", result[0])
print("Valor p:", result[1])
print("Número de lags usados:", result[2])
print("Número de observaciones usadas:", result[3])
print("Valores críticos:")
for key, value in result[4].items():
    print(f"   {key}: {value}")

# Interpretación
if result[1] <= 0.05:
    print("\nLa serie es estacionaria (rechazamos la hipótesis nula).")
else:
    print("\nLa serie no es estacionaria (no rechazamos la hipótesis nula).")
```

Dado que el estadístico ADF es mucho menor que los valores críticos y el valor p es significativamente menor que 0.05, podemos rechazar la hipótesis nula. Esto indica que la serie es estacionaria.

Por lo tanto, no necesitas aplicar transformaciones adicionales (como diferenciación) para hacerla estacionaria antes de ajustar tu modelo de regresión. Puedes proceder con el modelado sabiendo que la serie no tiene tendencia ni estacionalidad que deban ser corregidas.

## 4.1 Primer modelo

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Seleccionar características y variable objetivo
features = ['Lag_1', 'Lag_7', 'Date', 'Month', 'Media_Movil']
X = df_22_23[features]  # Características
y = df_22_23['Daily AQI Value']     # Variable objetivo

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.01, random_state=42)

# Crear el modelo XGBoost
model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1)

# Ajustar el modelo
model.fit(X_train, y_train)

# Hacer predicciones
y_pred = model.predict(X_test)
```

```python
# Error Cuadrático Medio (MSE)
mse = mean_squared_error(y_test, y_pred)
print(f"Error Cuadrático Medio (MSE): {mse}")

# Raíz del Error Cuadrático Medio (RMSE)
rmse = np.sqrt(mse)
print(f"Raíz del Error Cuadrático Medio (RMSE): {rmse}")

# Error Absoluto Medio (MAE)
mae = mean_absolute_error(y_test, y_pred)
print(f"Error Absoluto Medio (MAE): {mae}")

# Coeficiente de Determinación (R²)
r2 = r2_score(y_test, y_pred)
print(f"Coeficiente de Determinación (R²): {r2}")

# Error Absoluto Medio Percentil (MAPE)
# Para evitar ladivisión por cero
non_zero_indices = y_test != 0
mape = np.mean(np.abs((y_test[non_zero_indices] - y_pred[non_zero_indices]) / y_test[non_zero_indices])) * 100
print(f"Error Absoluto Medio Percentil (MAPE): {mape}%")
```

## 4.2 Segundo modelo

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split, GridSearchCV

# Seleccionar características y variable objetivo
features = ['Lag_1', 'Lag_7', 'Date', 'Month', 'Media_Movil']
X = df_22_23[features]  # Características
y = df_22_23['Daily AQI Value']  # Variable objetivo

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.01, random_state=42)

# Definir el espacio de parámetros para GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 150],  # Número de árboles
    'learning_rate': [0.01, 0.1, 0.2],  # Tasa de aprendizaje
    'max_depth': [3, 5, 7],  # Profundidad máxima de los árboles
    'min_child_weight': [1, 3, 5],  # Peso mínimo de los hijos
    'subsample': [0.7, 0.8, 0.9],
    'colsample_bytree': [0.7, 0.8, 0.9]
}

# Crear el modelo XGBoost
model = xgb.XGBRegressor(objective='reg:squarederror')

# Configurar GridSearchCV
grid_search = GridSearchCV(estimator=model, param_grid=param_grid,
                           scoring='neg_mean_squared_error',
                           cv=3, verbose=1, n_jobs=-1)

# Ajustar GridSearchCV
grid_search.fit(X_train, y_train)

# Obtener el mejor modelo y los mejores parámetros
best_model = grid_search.best_estimator_
best_params = grid_search.best_params_
best_score = grid_search.best_score_

# Hacer predicciones con el mejor modelo
y_pred = best_model.predict(X_test)
```
